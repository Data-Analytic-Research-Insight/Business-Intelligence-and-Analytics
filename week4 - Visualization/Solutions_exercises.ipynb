```
import pandas as pd #nothing new here
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```
url = 'https://raw.githubusercontent.com/michalis0/Business-Intelligence-and-Analytics/master/data/Credit.csv'
data = pd.read_csv(url)
```

Correction of exercise 1 from the waltkthrough

```
data_bar = (data.groupby('Student')['Married'].value_counts())
data2 = data_bar[[2,3]]
graph = data2.plot(kind = 'bar')
graph.set_title('Number of Student married', size=18)
graph.set_xlabel('')
graph.set_ylabel('Number of people', size=14)
data_bar.head()
```

Correction of exercise 2 from the waltkthrough

```
data_pie = data['Married'].value_counts()
graph = data_pie.plot(kind = 'pie', figsize=(7,5),autopct='%.2f%%')
graph.set_title("Proportion of gender", size=18)
```

Correction of exercise 3 from the waltkthrough

```
#Exemple de code
heat_map = data[['Limit','Income', 'Age', 'Rating']].corr()
sns.heatmap(data = heat_map,cbar_kws = dict(use_gridspec=False,orientation="horizontal", label='Correlations'), annot= True, linewidths= .5, cmap= 'RdBu')
```
